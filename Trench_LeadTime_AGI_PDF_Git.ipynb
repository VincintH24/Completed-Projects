```python
import openai
import pdfplumber
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from textwrap import wrap
#OpenAI API key
openai.api_key = ''
# Define function to analyze text using GPT-3
def analyze_text(text, page_title):
    # Include specific objectives within the prompt as we must COMMAND our artificial friends effectively
    prompt = (
        f"Please analyze the following text for '{page_title}':\n"
        f"1. Identify the lead times mentioned in weeks.\n"
        f"2. Provide details about the objects or processes associated with each lead time.\n"
        f"\n{text}"
    )
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=prompt,
        max_tokens=150,
        n=1
    )
    return response.choices[0].text.strip()
# PDF file paths
pdf_files = [
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_December-2022_Rev-2 (1).pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_February-2023.pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_March-2023_Rev-1 (1).pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_April-2023.pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_May-2023 (1).pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_June-2023.pdf",
    r"C:\Users\daddy\Downloads\New Compressed (zipped) Folder\Trench Lead Time Advisory_July-2023.pdf",
]
logo_path1 = r"C:\Users\daddy\Downloads\NiSource Supply Chain Market Insights.pptx.png"
# Define a function to add the logo as the first page
def add_logo_as_first_page(logo_path, pdf_output):
    fig, ax = plt.subplots(figsize=(16, 8.5))
    logo_image = plt.imread(logo_path)
    ax.imshow(logo_image, extent=[0, 8.5, 0, 11])
    ax.axis('off')
    pdf_output.savefig(fig, bbox_inches='tight')
    plt.close() 
# Define a function to add the Table of Contents as the second page
def add_table_of_contents(pdf_output):
    plt.figure(figsize=(11, 8))
    toc_text = (
        "Table of Contents\n\n"
        "Page 2: Table Of Contents\n"
        "Page 3: AI Analysis for Trench Lead Time Advisories December, 2022\n"
        "Page 4: AI Analysis for Trench Lead Time Advisories February, 2023\n"
        "Page 5: AI Analysis for Trench Lead Time Advisories March, 2023\n"
        "Page 6: AI Analysis for Trench Lead Time Advisories April, 2023\n"
        "Page 7: AI Analysis for Trench Lead Time Advisories May, 2023\n"
        "Page 8: AI Analysis for Trench Lead Time Advisories June, 2023\n"
        "Page 9: AI Analysis for Trench Lead Time Advisories July, 2023"
    )
    plt.text(0, 1, toc_text, va='top', ha='left', fontsize=14, wrap=True)
    plt.axis('off')
    pdf_output.savefig()
    plt.close() 
# Define a function to analyze each PDF and create a new PDF with the gpt-3 insights
def analyze_and_create_pdf(pdf_files):
    with PdfPages("insights.pdf") as pdf_output:
        # Add the logo as the first page
        add_logo_as_first_page(logo_path1, pdf_output)
        # Add the Table of Contents as the second page
        add_table_of_contents(pdf_output)
        page_titles = [
            "Trench Lead Time Advisories December, 2022",
            "Trench Lead Time Advisories February, 2023",
            "Trench Lead Time Advisories March, 2023",
            "Trench Lead Time Advisories April, 2023",
            "Trench Lead Time Advisories May, 2023",
            "Trench Lead Time Advisories June, 2023",
            "Trench Lead Time Advisories July, 2023"
        ]
        for pdf_file, page_title in zip(pdf_files, page_titles):
            pdf = pdfplumber.open(pdf_file)
            pdf_text = ""
            # Create a new page for each page in the PDF
            for page in pdf.pages:
                pdf_text += page.extract_text()
            insights = analyze_text(pdf_text, page_title)
            # Split the AI insights into sentences and format as bullet points with spacing to ensure a clean output homies
            sentences = insights.split('. ')
            bullet_points = "\n\n".join([f"â€¢ {sentence}" for sentence in sentences])
            # Create a new page for each PDF with the bullet points of the AI insights centered
            fig, ax = plt.subplots(figsize=(11, 8.5))
            ax.text(0.5, 0.5, bullet_points, va='center', ha='center', fontsize=12, wrap=True)
            ax.set_xlim([0, 8.5])
            ax.set_ylim([0, 11])
            ax.axis('off')
            pdf_output.savefig(fig, bbox_inches='tight')
            plt.close()
            pdf.close()
# Call the function to analyze and create the new PDF
analyze_and_create_pdf(pdf_files)
```
